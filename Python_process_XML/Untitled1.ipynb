```python
def walkData(root,ser):
    temp_list = [root.tag,root.attrib]
    if(root.text and (not re.match("\n",root.text))):
        temp_list.append(root.text)
    else:
        temp_list.append(" ")
    
    ser.append(temp_list)
    children_node = list(root)
    if len(children_node) == 0:
        return
    for child in children_node:
        walkData(child, ser)
    return
resultDataFrame = []
resultDataFeame2 = pd.DataFrame([0],columns=["PMID"])
flag = 0
date = {"Year":1,"Month":2,"Day":3,"Hour":4,"Minute":5}
for PubmedArticle in root:
    flag+=1
    count = 1
    ser = []
    walkData(PubmedArticle,ser)
    tempser = []
    columns = []
    data = {}
    for ind,i in enumerate(ser):
        if(i[0] in date.keys()):
            key = ser[ind-date[i[0]]][0]+"_"+i[0]
            if(key not in columns):
                columns.append(key)
                data[key] = []
                data[key].append(i[2])
            else:
                data[key].append(i[2])
            continue
        if(len(i[1])==0 and i[2]==" "):
            pass
        else:
            if(len(i[1])==0):
                if(i[0] not in columns):
                    columns.append(i[0])
                    data[i[0]] = []
                    data[i[0]].append(i[2])
                else:
#                     columns.append(i[0]+str(count))
#                     count+=1
                    data[i[0]].append(i[2])
#                 data.append(i[2])
            elif(i[2]==" "):
                for key in i[1].keys():
                    if(i[0]+"_"+key not in columns):
                        columns.append(i[0]+"_"+key)
                        data[i[0]+"_"+key] = []
                        data[i[0]+"_"+key].append(i[1][key])
                    else:
#                         columns.append(i[0]+"_"+key+str(count))
#                         count+=1
                        data[i[0]+"_"+key].append(i[1][key])
#                     data.append(i[1][key])
            else:
                if(i[0] not in columns):
                    columns.append(i[0])
                    data[i[0]] = []
                    data[i[0]].append(i[2])
                else:
#                     columns.append(i[0]+str(count))
#                     count+=1
                    data[i[0]].append(i[2])
#                 data.append(i[2])
                for key in i[1].keys():
                    if(i[0]+"_"+key not in columns):
                        columns.append(i[0]+"_"+key)
                        data[i[0]+"_"+key] = []
                        data[i[0]+"_"+key].append(i[1][key])
                    else:
#                         columns.append(i[0]+"_"+key+str(count))
#                         count+=1
                        data[i[0]+"_"+key].append(i[1][key])
#                     data.append(i[1][key])
            tempser.append(i)
#     print(len(ser))
#     print(len(tempser),end=" ")
#     print(len(columns),end=" ")
#     print(len(data),end=" ")
#     print(columns)
#     dat = dict(zip(columns,data))
#     s = pd.DataFrame(dat,index=[0])
#     break
    temp = {}
    for k,v in data.items():
        if(len(v)>1):
            temp[k] = [v]
        else:
            temp[k] = v
    s = pd.DataFrame(temp)
#     print(s.shape,end=" ")
    resultDataFeame2 = pd.concat([resultDataFeame2,s])
#     print(resultDataFeame2.shape)
#     if(flag==30):
#         break
#     resultDataFrame.append(s)
#     print(len(resultDataFrame))
    if(flag%100==0):
        print(flag)
        print(resultDataFeame2.shape)
```
